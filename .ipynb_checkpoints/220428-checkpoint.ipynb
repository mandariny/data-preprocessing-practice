### 서울시 집값 평균, 젤 비싼데

```python
df['지역2'] = df['시군구'].apply(lambda x : ' '.join(x.split(" ")[:2]))
```

```python
df.head()
```

```python
df_price = df[df['지역2'].apply(lambda x : x.split(" ")[0]) == '서울특별시'][['지역2', '단지명', '거래금액(만원)']]
```

```python
df_price.head()
```

```python
df_price['지역'] = df_price['지역2'].apply(lambda x : x.split(" ")[1])
```

```python
df_price.head()
```

```python
df_price.sort_values(by="거래금액(만원)").groupby(['지역'])[['거래금액(만원)']].first()
```

```python
df_price['거래금액(만원)'] = df_price['거래금액(만원)'].apply(lambda x : int(x.replace(",", "")))
```

```python
df_price.drop(['단지명', '지역2'], axis=1).groupby(['지역'])[['거래금액(만원)']].mean().sort_values(by="거래금액(만원)", ascending=False)
```

# 강남구 월별 평균 매매가격 추이(그래프)

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['계약년월'].unique()
```

```python
df_gn = df[df['지역2'].apply(lambda x : x.split(" ")[1]) == '강남구'][['지역2', '계약년월', '거래금액(만원)']]
```

```python
df_gn.columns=['지역', '월', '거래금액']
```

```python
df_gn['거래금액'] = df_gn['거래금액'].apply(lambda x : int(x.replace(",", "")))
```

```python
df_gn.head()
```

```python
df_gn = df_gn.groupby(["월"], as_index=False)[['거래금액']].mean()
```

```python
df_gn.head()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(15, 10))
plt.plot(df_gn['월'], df_gn['거래금액'])

plt.title("Gangnam Apart Price")
plt.xlabel("Month")
plt.ylabel("Price")

plt.show()
```
