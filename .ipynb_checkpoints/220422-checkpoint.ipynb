## 12월 한 달 평균 가장 휘발유 가격이 비싼 동네는?(지역구 기준)

```python
total.head()
```

```python
total['기간'].unique()
```

```python
total['월'] = total['기간'].apply(lambda x : str(x)[4:6])
```

```python
total.tail()
```

```python
df = total.loc[total['월'] == '12', ['지역', '월', '휘발유']]
```

```python
df.head()
```

```python
df.groupby(['지역'], as_index=False)[['휘발유']].mean().sort_values(by='휘발유', ascending=False)
```

## 풀이

```python
df_201912 = total[total['기간'].apply(lambda x : str(x)[:6] == '201912')].copy()
```

```python
df_201912_mean = df_201912.groupby("지역")[['휘발유']].mean().sort_values(by=['휘발유'], ascending=False)
```

```python
df_201912_mean[df_201912_mean['휘발유'] > df_201912_mean.mean().values[0]]
```
