# 문제1

```python
import pandas as pd
```

```python
pd. set_option('display.max_columns', 500)
```

```python
column = pd.read_csv('./data/Subway/Subway_201607.csv', encoding='cp949').columns
```

```python
column
```

```python
df_all = pd.DataFrame(columns=column)
for y in ["{}".format(x).zfill(2) for x in range(1, 13)]:
    df = pd.read_csv(f'./data/Subway/Subway_2016{y}.csv', encoding='cp949')
    
    if '04 ~ 05' in df.columns : 
        df = df.drop(['04 ~ 05', '02 ~ 03', '03 ~ 04'], axis=1)
        
    df.rename(columns = lambda x: x.replace("시", "").replace(" ", ""), inplace = True)  
    df_all = pd.concat([df_all, df])
```

```python
df_all = df_all.drop(["호선", "할인"], axis=1)
```
